```python
import pandas as pd
import numpy as np
import lib_tools as pt
from datetime import date

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

dic_usagers = pt.load_usagers(2020, 2021)
dic_caract  = pt.load_caract(2020, 2021)
dic_vehic   = pt.load_vehicules(2020, 2021)
dic_lieux   = pt.load_lieux(2020, 2021)

df_usagers  = dic_usagers[2020]
df_caract   = dic_caract[2020]
df_vehic    = dic_vehic[2020]
df_lieux    = dic_lieux[2020]
```

## Intercorrelations des variables de chaque table

```python
cor_caract = df_caract[['lum', 'dep', 'com', 'agg', 'int', 'atm', 'col', 'adr', 'lat', 'long']].corr()
cor_usagers = df_usagers[['place', 'catu', 'grav', 'sexe', 'an_nais', 'trajet', 'secu1', 'secu2', 'secu3', 
                  'locp', 'actp', 'etatp']].corr()
cor_vehic = df_vehic[['senc', 'catv', 'obs', 'obsm', 'choc', 'manv', 'motor', 'occutc']].corr()
cor_lieux = df_lieux[['catr', 'circ', 'nbv', 'vosp', 'prof', 'plan', 'lartpc', 'larrout', 'surf', 'infra', 'situ','vma']].corr()

# fig, ax = plt.subplots(figsize=(5,5))
fig, ((ax1, ax2) , (ax3, ax4)) = plt.subplots(2,2)
fig.set_size_inches(12, 12)

ax1.set_title('Intercorrélations des var de caractéristiques')
sns.heatmap(cor_caract, annot=True, ax=ax1, cmap='coolwarm');

ax2.set_title('Intercorrélations des var de usagers')
sns.heatmap(cor_usagers, annot=True, ax=ax2, cmap='coolwarm');

ax3.set_title('Intercorrélations des var de véhicules')
sns.heatmap(cor_vehic, annot=True, ax=ax3, cmap='coolwarm');

ax4.set_title('Intercorrélations des var de lieux')
sns.heatmap(cor_lieux, annot=True, ax=ax4, cmap='coolwarm');
```
